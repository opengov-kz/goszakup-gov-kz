```python
import requests
import pandas as pd

api_url = "https://ows.goszakup.gov.kz/v3/refs/ref_pln_point_status"
token = "d5c3d78fc111d88a0a37b4ab8f83cbd5"
headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}
file = "plan_point_status.csv"

def retrieve_plan_point_status():
    """Получает статусы пунктов планов через API и сохраняет в CSV."""
    status_list = []

    try:
        response = requests.get(api_url, headers=headers, timeout=10)
        response.raise_for_status()
        response_data = response.json()

        records = response_data.get("items", [])
        if not records:
            print("Нет данных для загрузки.")
            return

        for record in records:
            status_list.append({
                "id": record.get("id"),
                "code": record.get("code"),
                "name_kz": record.get("name_kz"),
                "name_ru": record.get("name_ru")
            })

        print(f"Загружено записей: {len(status_list)}")
        export_to_csv(status_list)

    except requests.exceptions.RequestException as err:
        print(f"Ошибка запроса: {err}")

def export_to_csv(dataset):
    """Записывает данные в CSV-файл."""
    if not dataset:
        print("Нет данных для экспорта.")
        return

    data_frame = pd.DataFrame(dataset)
    data_frame.to_csv(
        file,
        index=False,
        encoding="utf-8-sig",
        sep="|",
        quotechar="'",
        escapechar="\\"
    )
    print(f"Данные сохранены в {file}")

retrieve_plan_point_status()
```
